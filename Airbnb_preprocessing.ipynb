```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("Belgium.csv")
df1
```

```python
df2=pd.read_csv("France.csv")
df2
```

```python
df3=pd.read_csv("Germany.csv")
df3
```

```python
df4=pd.read_csv("Seattle.csv")
df4
```

```python
# Standardizing column names (converting to lowercase for consistency and adding _ on space so that all columns name is same)
df1.columns = df1.columns.str.lower().str.replace(' ', '_')
df2.columns = df2.columns.str.lower().str.replace(' ', '_')
df3.columns = df3.columns.str.lower().str.replace(' ', '_')
df4.columns = df4.columns.str.lower().str.replace(' ', '_')
```

```python
# Add missing 'country' column to Belgium, France, and Germany datasets
df1['country'] = 'Belgium'
df2['country'] = 'France'
df3['country'] = 'Germany' #seattle already has country column US so adding country column for remaining data set

# Concatenate all datasets
df_final=pd.concat([df1,df2,df3,df4], ignore_index=True)

# Save merged dataset
#df_final.to_csv('airbnb_merged.csv', index=False)

print("Datasets merged successfully!")
```

```python
#Convert host_since to Date Format
#Converts invalid dates to NaT (Not a Time) instead of crashing.
df_final['host_since'] = pd.to_datetime(df_final['host_since'], errors='coerce')
# Check if it worked
print(df_final['host_since'].head())
```

```python
#Convert host_response_rate to a Number
df_final['host_response_rate'] = df_final['host_response_rate'].astype(str)  # Convert all values to string
df_final['host_response_rate'] = df_final['host_response_rate'].str.replace('%', '', regex=True)  # Remove '%'
df_final['host_response_rate'] = pd.to_numeric(df_final['host_response_rate'], errors='coerce')  # Convert to float

# Check the result
print(df_final['host_response_rate'].head())
```

```python
#Convert Price Columns to Numbers
#price, weekly_price, monthly_price, and security_deposit contain comma signs (,).
# List of price-related columns
price_columns=['price','weekly_price','monthly_price','security_deposit']

# Loop through each column and clean it
for col in price_columns:
    df_final[col] = df_final[col].astype(str)  # Convert to string (just in case)
    df_final[col] = df_final[col].str.replace(',', '')  # Remove commas
    df_final[col] = pd.to_numeric(df_final[col], errors='coerce')  # Convert to numbers

# Check if conversion worked
print(df_final[price_columns].head())
```

```python
#Here Zip Code has both format numeric and string so based on occurence chnaging
df_final['zipcode'] = df_final['zipcode'].astype(str).str.strip().str.upper()
df_final['zipcode'] = df_final['zipcode'].str.replace(r'[^A-Za-z0-9]', '', regex=True)
df_final['zipcode'] = df_final['zipcode'].fillna(df_final['zipcode'].mode()[0])  #'75018' is a mode value
```

```python
# Function to clean special characters
import re  #Regular Expression module, used for text matching and manipulation
def clean_column(value):
    if pd.isnull(value):
        return 'Not Available'
    # Replace if special characters are detected
    if re.search(r'[^a-zA-Z0-9\s]', str(value)):
        return 'Not Available'
    return value

# Apply cleaning for host_name, city, and state columns
columns_to_clean=['host_name', 'city', 'state']
for col in columns_to_clean:
    df_final[col]=df_final[col].apply(clean_column)
```

```python
#Treating missing values for categorical column
# Fill with "Unknown" for identifiers
df_final['host_name'] = df_final['host_name'].fillna("NA")

# Fill with mode (most frequent value)
df_final['host_since'] = df_final['host_since'].fillna(df_final['host_since'].mode()[0])
df_final['city'] = df_final['city'].fillna(df_final['city'].mode()[0])
df_final['state'] = df_final['state'].fillna(df_final['state'].mode()[0])
df_final['property_type'] = df_final['property_type'].fillna(df_final['property_type'].mode()[0])

# Fill with default text for amenities
df_final['amenities'] = df_final['amenities'].fillna("No Amenities")
```

```python
#Treating missing vale for numerical 
# Host response rate (percentage) - Fill with median
df_final['host_response_rate']=df_final['host_response_rate'].fillna(df_final['host_response_rate'].median())

# Host listing counts - Fill with median
df_final['host_listings_count']=df_final['host_listings_count'].fillna(df_final['host_listings_count'].median())
df_final['host_total_listings_count']=df_final['host_total_listings_count'].fillna(df_final['host_total_listings_count'].median())

# Property attributes - Fill with median
df_final['bathrooms']=df_final['bathrooms'].fillna(df_final['bathrooms'].median())
df_final['bedrooms']=df_final['bedrooms'].fillna(df_final['bedrooms'].median())
df_final['beds']=df_final['beds'].fillna(df_final['beds'].median())

# Price-related columns - Fill missing price, estimate weekly/monthly price
df_final['price']=df_final['price'].fillna(df_final['price'].median())
df_final['weekly_price']=df_final['weekly_price'].fillna(df_final['price'] * 7)
df_final['monthly_price']=df_final['monthly_price'].fillna(df_final['price'] * 30)

# Security deposit & cleaning fee - Fill missing values with 0
df_final['security_deposit']=df_final['security_deposit'].fillna(0)
df_final['cleaning_fee']=df_final['cleaning_fee'].fillna(0)

# Review-related columns - Fill missing values
df_final['review_scores_rating']=df_final['review_scores_rating'].fillna(df_final['review_scores_rating'].median())
df_final['reviews_per_month']=df_final['reviews_per_month'].fillna(0) 
```
